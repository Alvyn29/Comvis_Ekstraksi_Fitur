# **3. FUNCTION EKSTRAKSI RGB + HSV**

```python
def extract_features(image_path):
    img = cv2.imread(image_path)

    # Resize biar konsisten
    img = cv2.resize(img, (256, 256))

    # RGB
    rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # HSV
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    # Flatten
    rgb_pixels = rgb.reshape(-1, 3)
    hsv_pixels = hsv.reshape(-1, 3)

    # ======================
    # RGB Features
    # ======================
    rgb_mean = np.mean(rgb_pixels, axis=0)
    rgb_std = np.std(rgb_pixels, axis=0)

    # ======================
    # HSV Features
    # ======================
    hsv_mean = np.mean(hsv_pixels, axis=0)

    # Brightness (V channel)
    brightness = np.mean(hsv[:,:,2])

    # Jumlah pixel
    total_pixel = rgb.shape[0] * rgb.shape[1]

    # Gabungkan fitur
    features = np.concatenate([
        rgb_mean, rgb_std,
        hsv_mean,
        [brightness],
        [total_pixel]
    ])

    return features
```

# **4. LOOP DATASET**

```python
data = []

for file_name in os.listdir(folder_path):
    if file_name.endswith(('.jpg', '.png', '.jpeg')):

        img_path = os.path.join(folder_path, file_name)

        fitur = extract_features(img_path)

        data.append([file_name] + fitur.tolist())
```

# **5. Membuat Dataset (DataFrame)**

```python
columns = [
    'Nama File',

    'Rata-rata Merah', 'Rata-rata Hijau', 'Rata-rata Biru',
    'Variasi Merah', 'Variasi Hijau', 'Variasi Biru',

    'Rata-rata Hue', 'Rata-rata Saturasi', 'Rata-rata Value',

    'Tingkat Kecerahan',
    'Jumlah Pixel'
]

df = pd.DataFrame(data, columns=columns)

df
```
